```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var1=[1,2,3,4,5]
var2=[3,4,5,2,6]
plt.plot(var1,var2)         #plot linegraph using matplotlib
plt.show()
```

```python
sns.lineplot(x=var1,y=var2)             #plotting same graph using seaborn
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'var1':var1,'var2':var2})      #making dataframe in pandas
df
```

```python
# now plotting the same graph using seaborn
sns.lineplot(x='var1',y='var2',data=df)     #seaborn gives upgraded functionality than matplotlib
plt.show()                                  # here it gives label in x-axis and y-axis
```

```python
df1=sns.load_dataset("penguins").head(20)
df1
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue="sex")   # heu divide graph into parts based on gender 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex')
plt.grid()
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex')
plt.title("penguin plot ")
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',style='sex',markers=["o","<"])
plt.grid()
plt.title("penguin plot")
plt.show()
```

```python
sns.histplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex')
plt.grid()      #to make histogram plot
plt.show()
```

```python
df2=sns.load_dataset("penguins")
df2
```

```python
df2['island']
```

```python

sns.barplot(x=df2.island,y=	df2.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df2,hue='sex',palette="Accent")
plt.show()          #palette gives the different color to bargraph
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df2,hue='sex')
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df2,hue='sex',hue_order=['Female','Male'])
plt.show()      #heu_orded change the order
```

```python
sns.displot(df2["flipper_length_mm"])
plt.show()          # didplot makes the histogram
```

```python
sns.displot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()                          #bins gives the proper order betweens the intervals
```

```python
sns.displot(df2['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,color='red')
plt.show()                          #kde gives the line in the histogram
```
